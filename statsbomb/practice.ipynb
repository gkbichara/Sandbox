```python
import pandas as pd
import requests

# URL for La Liga 2020/2021 matches
url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/11/90.json"

# Fetch data
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    matches = pd.DataFrame(response.json())  # Convert JSON to DataFrame
    print(matches.head())  # Display first few rows
else:
    print("Failed to fetch data. Status code:", response.status_code)
```
